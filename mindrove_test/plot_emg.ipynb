# Task 5
Sacar el valor medio de cada EMG para cada task, para cada repeticion

```python
from statistics import mean
import numpy as np
import statistics
id = 1
dataval= dataframe[id]
emgval = dataval.iloc[:,2:10] * 0.045
numemg = [1,2,3,4,5,6,7,8]
tasks = [1,2,3,4,5,6,7]
tasksnames = ['Thumb flexion', 'Index finger flexion','Middle finger flexion','Ring finger flexion', 'Pinky finger flexion','Wrist extension (outward)','Wrist flexion (inward)' ]
taskcolumn = 1
taskval = dataval.iloc[:,taskcolumn]
task = np.zeros((3,8))
#new dataframe contaning all the media, for all sucjects, for all tasks, for each rep
dataframetaskmedia = []
indtask = []
for m in dataframe:
    for k,val in enumerate(tasks):
        title = ('TASK NUMBER #', val, ':', tasksnames[k])
        rep = 0
        start = 0
        tasksaved = 0
        for j in range(len(taskval)):
            if taskval[j] == tasks[k] and taskval[j-1] == 0:
                start = j
            elif taskval[j] == tasks[k] and taskval[j + 1] == 0:
                count = 0
                for i in numemg:
                    media = emgval.iloc[start:j,count]
                    task[rep,count] = statistics.mean(media)
                    count = count + 1
                rep = rep + 1
                start = 0
        indexs = ['rep 1','rep 2','rep 3' ]
        #tasksaved: save each task result for each subject
        tasksaved = pd.DataFrame(data = task, index = indexs, columns=titles)
        #tasksaved.rename_axis(["TASK NUMBER #", val, ':', tasksnames[k]], axis="columns", inplace=True), me gustaria poner un titulo que cambie por task
    #indtask: save for each subject the whole tasks results
    indtask.append(tasksaved)
    
 #dataframetaskmedia: save all    
dataframetaskmedia.append(indtask)
print(indtask)







    
    

```
